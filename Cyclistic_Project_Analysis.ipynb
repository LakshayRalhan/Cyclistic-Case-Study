```python
import pandas as pd
```

```python
df = pd.read_csv('cyclistic_cleaned.csv')
```

```python
df.head()
```

```python
#calculating avg ride length for each rider type

df.groupby('member_casual')['ride_length'].mean()
```

```python
#total number of rides by each user type

df['member_casual'].value_counts()
```

```python
# Define correct order of days of week instead of alphabetical (default)

day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

df['day_of_week'] = pd.Categorical(df['day_of_week'], categories = day_order)

#Define correct order of month

month_order = ['January','February','March','April','May','June','July','September','October','November','December']

df['month'] = pd.Categorical(df['month'], categories = month_order)
```

```python
#count rides by day of week and rider type

rides_by_day = df.groupby(['day_of_week', 'member_casual']).size()

rides_by_day.unstack()
```

```python
#calculate avg ride length for each day and rider type

avg_length_day = df.groupby(['day_of_week','member_casual'])['ride_length'].mean()

avg_length_day.unstack()
```

```python
#count rides by month and rider type

rides_by_month = df.groupby(['month', 'member_casual']).size()

rides_by_month.unstack()
```

```python
#calculate avg ride length by rider type and bike type

bike_usage = df.groupby(['rideable_type','member_casual'])['ride_length'].mean()

bike_usage.unstack()
```

```python
#count rides by hour and rider type

rides_by_hour = df.groupby(['hour','member_casual']).size()

rides_by_hour.unstack()
```

```python
#INSIGHTS
# 1. Members take more trips but casual riders take longer trips on average.
# 2. Casual riders ride significantly more on weekends, while Members ride more during weekdays.
# 3. Casual riders also take longer rides on weekends, while the average ride length for Members is consistent throughout the week.
# 4. Bike usage increase during the summer months for both Casual riders and Members.
# 5. Classic bikes are preferred over electric bikes by both Members and Casual riders.
# 6. Members ride more during commuting hours, while Casual riders ride more later during the day
```
