```python
cyclistic = pd.concat([jan, feb, mar, apr, may, jun, jul, aug, sep, octo, nov, dec], ignore_index = True)
```

```python
#check for number for null values in each column
cyclistic.isnull().sum()
```

```python
#check if duplicated rows exist in the dataframe
cyclistic.duplicated().sum()
```

```python
#convert started_at and ended_at columns to datetime format
cyclistic['started_at'] = pd.to_datetime(cyclistic['started_at'])
cyclistic['ended_at'] = pd.to_datetime(cyclistic['ended_at'])

#verify data types
cyclistic.dtypes
```

```python
#add ride_length column
cyclistic['ride_length'] = cyclistic['ended_at'] - cyclistic['started_at']

#convert ride_length values to minutes
cyclistic['ride_length'] = cyclistic['ride_length'].dt.total_seconds() / 60
```

```python
#Rides less than 1 minute could be errors or test rides. So we remove them.
cyclistic = cyclistic[cyclistic['ride_length'] > 1]
```

```python
#remove duplicate rides based on ride_id
cyclistic = cyclistic.drop_duplicates(subset = 'ride_id')
```

```python
#check again for null values after removing duplicates
cyclistic.isnull().sum()
```

```python
#remove rows with missing start or end station names.
cyclistic = cyclistic.dropna(subset = ['start_station_name','end_station_name'])
```

```python
#CREATE NEW COLUMNS FOR ANALYSIS
```

```python
#extract the day of the week from started_at column to analyze weekday vs weekend behavior
cyclistic['day_of_week'] = cyclistic['started_at'].dt.day_name()

#extract month name to monitor monthly trends
cyclistic['month'] = cyclistic['started_at'].dt.month_name()

#extract hour of the ride start time to identify peak riding hours during the day.
cyclistic['hour'] = cyclistic['started_at'].dt.hour
```
